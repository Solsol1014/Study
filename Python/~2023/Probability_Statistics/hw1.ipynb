```python
import matplotlib.pyplot as plt
import random
```

```python
def g(x):
    if x>0.9:
        return 1
    elif x>0.8:
        return 2
    elif x>0.6:
        return 3
    elif x>0.3:
        return 4
    else:
        return 5
```

```python
y = []
for i in range(1, 100000):
    y.append(g(random.random()))

#print(y)
plt.hist(y, bins=5)
plt.grid(True)
plt.show()
```
