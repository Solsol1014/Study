```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm, expon
```

```python
def cdf(n, sample_size, expon = 0, lambd = 0.0):
    mean = 5
    var = 100/12
    if expon==0:
        X = np.random.uniform(0, 10, (n, sample_size))
        Sn = np.sum(X, axis=0)
        Zn = (Sn-n*mean)/(np.sqrt(n)*np.sqrt(var))
    else:
        X = np.random.exponential(scale = 1/lambd, size=(n, sample_size))
        Sn = np.sum(X, axis=0)
        Zn = (Sn-n/lambd)/np.sqrt(n/(lambd**2))

    x = np.sort(Zn)
    y = np.arange(1, sample_size+1)/sample_size

    plt.plot(x, y, label='n = {} empirical CDF'.format(n))

    return x
```

```python
n = 1
sample_size = 10000

x = cdf(n, sample_size)

normcdf = norm.cdf(x)

plt.plot(x, normcdf, label='Φ(z)')

plt.xlabel('Z')
plt.ylabel('CDF')
plt.title('Empirical CDF vs Standard Normal CDF')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(40, 30))
sample_size = 10000
ns = [1, 2, 3, 5, 10, 100]

for n in ns:
    cdf(n, sample_size)

plt.xlabel('Z')
plt.ylabel('CDF')
plt.title('Graph of CDF as n grows')
plt.legend()
plt.show()
```

n이 커짐에 따라 empirical CDF가 Standard normal CDF에 더 가까워진다는 것을 확인할 수 있습니다.

```python
sample_size = 10000
lambd = 0.5

x = cdf(1, sample_size, 1, lambd)

normcdf = norm.cdf(x)

plt.plot(x, normcdf, label='Φ(z)')

plt.xlabel('Z')
plt.ylabel('CDF')
plt.title('Empirical CDF vs Standard Normal CDF (Exponential Random Variables)')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(40, 30))
sample_size = 10000
ns = [1, 5, 10, 100, 1000]

for n in ns:
    cdf(n, sample_size, 1, lambd)

plt.xlabel('Z')
plt.ylabel('CDF')
plt.title('Graph of CDF as n grows (Exponential Random Variables)')
plt.legend()
plt.show()
```

지수확률변수에서도 균일분포에서와 같이 n이 증가할수록 Empirical CDF가 Standard Normal CDF에 가까워진다는 사실을 알 수 있습니다.
