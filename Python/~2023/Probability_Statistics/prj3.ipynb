```python
import numpy as np
import matplotlib.pyplot as plt

def binomial_simulation(n, p, K):
    # (a) Draw K samples from binomial RV with parameters (n, p)
    samples = np.random.binomial(n, p, K)
    #print(samples)
    
    # (b) Plot the histogram of the samples
    plt.figure(figsize=(12, 10))
    plt.subplot(2, 2, 1)
    plt.hist(samples, density=True)
    plt.xlabel('Value')
    plt.ylabel('Probability')
    plt.title('Histogram of Samples')
    
    # (c) Plot the empirical CDF of the samples
    plt.subplot(2, 2, 2)
    sortsamples = np.sort(samples)
    #print(sortsamples)
    y = np.arange(1, K+1) / float(K)
    plt.plot(sortsamples, y)
    plt.xlabel('Value')
    plt.ylabel('Cumulative Probability')
    plt.title('Empirical CDF')
    
    # (d) Plot the PMF of the binomial RV with parameters (n, p)
    plt.subplot(2, 2, 3)
    x = np.arange(0, n+1)
    pmf = np.bincount(samples, minlength=n+1)/float(K)
    plt.bar(x, pmf)
    plt.xlabel('Value')
    plt.ylabel('Probability')
    plt.title('PMF of Binomial RV')
    
    # (e) Plot the CDF of the binomial RV with parameters (n, p)
    plt.subplot(2, 2, 4)
    cdf = np.cumsum(pmf)
    plt.plot(x, cdf)
    plt.xlabel('Value')
    plt.ylabel('Cumulative Probability')
    plt.title('CDF of Binomial RV')
    
    # Compare histogram and PMF
    plt.figure()
    plt.hist(samples, bins='auto', density=True, alpha=0.5, label='Histogram', color='lightcoral')
    plt.bar(x, pmf, alpha=0.5, label='PMF')
    plt.xlabel('Value')
    plt.ylabel('Probability')
    plt.title('Compare of Histogram and PMF')
    plt.legend()
    
    # Compare empirical CDF and CDF
    plt.figure()
    plt.plot(sortsamples, y, label='Empirical CDF')
    plt.plot(x, cdf, label='CDF')
    plt.xlabel('Value')
    plt.ylabel('Cumulative Probability')
    plt.title('Compare of Empirical CDF and CDF')
    plt.legend()
    
    plt.tight_layout()
    plt.show()

n = 50
p = 0.5
K = 1000

binomial_simulation(n, p, K)
```

```python
def exponential_simulation(lambd, K):
    # (a) Draw K samples from exponential RV with parameter lambda
    samples = np.random.exponential(scale=1.0/lambd, size=K)
    
    # (b) Plot the histogram of the samples
    plt.figure(figsize=(12, 10))
    plt.subplot(2, 2, 1)
    plt.hist(samples, bins='auto', density=True)
    plt.xlabel('Value')
    plt.ylabel('Probability')
    plt.title('Histogram of Samples')
    
    # (c) Plot the empirical CDF of the samples
    plt.subplot(2, 2, 2)
    sortsamples = np.sort(samples)
    y = np.arange(1, K+1) / float(K)
    plt.plot(sortsamples, y)
    plt.xlabel('Value')
    plt.ylabel('Cumulative Probability')
    plt.title('Empirical CDF')
    
    # (d) Plot the PDF of the exponential RV with parameter lambda
    plt.subplot(2, 2, 3)
    x = np.linspace(0, np.max(samples), 50)
    pdf = lambd * np.exp(-lambd * x)
    plt.plot(x, pdf)
    plt.xlabel('Value')
    plt.ylabel('Probability Density')
    plt.title('PDF of Exponential RV')
    
    # (e) Plot the CDF of the exponential RV with parameter lambda
    plt.subplot(2, 2, 4)
    cdf = 1 - np.exp(-lambd * x)
    plt.plot(x, cdf)
    plt.xlabel('Value')
    plt.ylabel('Cumulative Probability')
    plt.title('CDF of Exponential RV')
    
    # Compare histogram and PDF
    plt.figure()
    plt.hist(samples, bins='auto', density=True, alpha=0.5, label='Histogram')
    plt.plot(x, pdf, label='PDF')
    plt.xlabel('Value')
    plt.ylabel('Probability')
    plt.title('Compare of Histogram and PDF')
    plt.legend()
    
    # Compare empirical CDF and CDF
    plt.figure()
    plt.plot(sortsamples, y, label='Empirical CDF')
    plt.plot(x, cdf, label='CDF')
    plt.xlabel('Value')
    plt.ylabel('Cumulative Probability')
    plt.title('Compare of Empirical CDF and CDF')
    plt.legend()
    
    plt.tight_layout()
    plt.show()

lambd = 0.5
K = 1000

exponential_simulation(lambd, K)
```
